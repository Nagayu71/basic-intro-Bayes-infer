```python
# Code 1.1 くじ引き問題の確率分布の可視化

from scipy.special import comb
n = 5
x = range(n+1)
y = [comb(n ,i) / 2**n for i in x]
plt.bar(x, y)
plt.title("くじ引き問題の確率分布の可視化結果")
plt.show()
```

```python
# Code 1.2 くじ引き問題の確率分布の可視化

n = 1000
x = range(n+1)
y = [comb(n ,i) / 2**n for i in x]
plt.bar(x, y)
plt.xlim((430, 570))
plt.title("くじ引き問題の確率分布の可視化結果(n=1000)");
```

```python
# Code 1.3 確率分布と正規分布関数の重ね描き（中心極限定理の確認）

# 正規分布関数の定義
def norm(x, mu, sigma):
    return np.exp(-((x-mu)/sigma)**2 / 2) / (np.sqrt(2 * np.pi) * sigma)

n = 1000
# 二項分布のグラフ
plt.xlim((430, 570))
x = np.arange(430, 570)
y1 = [comb(n ,i) / 2**n for i in x]
plt.bar(x, y1)

# 正規分布のグラフ
mu = n/2
sigma = np.sqrt(mu/2)
y2 = norm(x, mu, sigma)
plt.plot(x, y2, c="k")

plt.title("確率分布と正規分布関数の重ね描き");
```

```python
# Figure 1.3 確率は面積である

n = 1000

# グラフ描画
plt.xlim((430,570))
x = np.arange(430, 571)
x1 = 460
x2 = 480
x_range = np.arange(x1, x2+1)

# 正規分布関数
mu = n/2
sigma = np.sqrt(mu/2)
plt.plot(x, norm(x, mu, sigma), c='k')
plt.fill_between(x_range, 0, norm(x_range, mu, sigma), facecolor='b', alpha=0.2) # 面積をハイライト
plt.plot((x1, x1), (0, norm(x1, mu, sigma)), c='b') # 区切り線
plt.plot((x2, x2), (0, norm(x2, mu, sigma)), c='b') # 区切り線
plt.title('正規分布関数と確率の関係');
```

```python
# Code 1.4 必要なライブラリのインポート

import pymc as pm
import arviz as az

print(f"Running on PyMC v{pm.__version__}")
print(f"Running on ArviZ v{az.__version__}")
```

```python
# Code 1.5 確率モデル定義

# 確率モデルコンテキスト用インスタンス
model = pm.Model()

# インスタンスをwith文で指定
with model:
    # Binomial: 二項分布の確率分布クラス
    # p: 二項分布の元になる試行の成功確率
    # n: 二項分布における試行回数
    # 'x':  確率変数 x のサンプル値を参照するときのラベル
    x = pm.Binomial("x", p=0.5, n=5)
```

```python
# Code 1.6 定義した確率モデルからサンプリング（事前分布）

"""
- with modelのコンテキスト定義により、上で定義した確率モデルと紐付けられる
- sample_prior_predictive: 事前分布の予測値取得関数
- 乱数により生成されたサンプル値（事前分布）が変数prior_samplesにセットされる
"""
with model:
    prior_samples = pm.sample_prior_predictive(random_seed=42)
```

```python
# Code 1.7 サンプリング結果を Notebook UI で確認

prior_samples
```

```python
# Code 1.8 （サンプリング結果の） NumPy 形式データの抽出と簡単な分析

# 事前分布としてのサンプル値を取得
x_samples = prior_samples["prior"]["x"].values
print("type: ", type(x_samples))
print("shape: ", x_samples.shape)
print("values: " , x_samples, '\n')

# 通常の NumPy データとして頻度分析をした例
value_counts = pd.DataFrame(x_samples.reshape(-1)).value_counts().sort_index()
print(value_counts)
```

```python
# Code 1.9 ArviZ によるサンプル値の集計

summary = az.summary(prior_samples, kind="stats")
display(summary)
```

```python
# Code 1.10 ArviZ によるサンプル値（事前分布）の可視化

ax = az.plot_dist(x_samples)
ax.set_title("ArviZによるサンプル値の可視化結果");
```
